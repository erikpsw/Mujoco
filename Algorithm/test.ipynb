```python
import kinematic
import numpy as np
import matplotlib.pyplot as plt

start_pos =np.array([0.29444878, 0, 0.0283401 ]) 
end_pos =  np.array([0.10699319124221803 ,0.005106544122099876 ,-0.1530148871243])
path_list=kinematic.generate_path(start_pos,end_pos,end_d=0.03)
result_list=kinematic.select_alpha(path_list,3.6756633967948966,np.pi*3/2)
```

```python
result_list=[res for res in result_list]
result_list
```

```python
angle_list=kinematic.inverse_kinematic_with_alpha(path_list,result_list)
```

```python


result_list=kinematic.normalize_to_pi(angle_list)
result_list=angle_list
lst=[]
for angle in result_list:
    lst.append(angle[0]+angle[1]+angle[2])

plt.plot(lst)
```
